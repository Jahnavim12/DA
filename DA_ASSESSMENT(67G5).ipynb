```python
brand_avg_prices = df.groupby('Brand')['Price'].mean().sort_values(ascending=False)
brand_avg_prices.count()
```

```python

d={}
for i in brand_avg_prices.index:
  print(i)
```

```python
d = {}
j=len(brand_avg_prices)
for i in brand_avg_prices.index:
  d[i] = j
  j-=1
d
```

```python
x=[]
for j in df['Brand']:
  y=d[j]
  x.append(y)
df['Encoded brand']=x
df
```

```python
fe=[]
min=df['Fuel Efficiency'].min()
max=df['Fuel Efficiency'].max()
for i in df['Fuel Efficiency']:
  scaled_value=(i-min)/(max-min)
  fe.append(scaled_value)
df['Scaled Fuel efficiency']=fe
df
```

```python
hp=[]
min=df['Horsepower'].min()
max=df['Horsepower'].max()
for i in df['Horsepower']:
  scaled_value=(i-min)/(max-min)
  hp.append(scaled_value)
df['Scaled Horsepower']=hp
df
```

```python
eb=[]
min=df['Encoded brand'].min()
max=df['Encoded brand'].max()
for i in df['Encoded brand']:
  scaled_value=(i-min)/(max-min)
  eb.append(scaled_value)
df['Scaled Encoded brand']=eb
df
```

```python
df['Performance Index']=(df['Horsepower'] * 0.7 + df['Fuel Efficiency'] * 0.3) / df['Price']
```
